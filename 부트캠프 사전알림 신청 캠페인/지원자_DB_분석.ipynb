나이를 나이대로 분류

```python
# 나이대 생성 함수 정의
def get_age(age):
    if age == 0:
        return '응답없음'            # 9999는 결측값 대체 → '응답없음'으로 분류
    if 10 <= age < 20:
        return '10대'
    elif 20 <= age < 30:
        return '20대'
    elif 30 <= age < 40:
        return '30대'
    elif 40 <= age < 50:
        return '40대'               # 40~49세 → 40대
    elif 50 <= age < 60:
        return '50대'               # 50~59세 → 50대
    elif 60 <= age:
        return '60대 이상'          # 60~99세 → 60대 이상
    else:
        return '응답없음'

# '나이' 컬럼에 함수 적용해서 '나이대' 컬럼 생성
dt['나이대'] = dt['나이'].apply(get_age)
dt['나이대'].value_counts().sort_index()
```

IQR

```python
# 직업별로 적용
def remove_outliers(df, group_col, target_col):
    cleaned_df = pd.DataFrame()

    for group, subset in df.groupby(group_col):
        Q1 = subset[target_col].quantile(0.25)
        Q3 = subset[target_col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # 이상치 제외
        cleaned_subset = subset[(subset[target_col] >= lower) & (subset[target_col] <= upper)]
        cleaned_df = pd.concat([cleaned_df, cleaned_subset])

    return cleaned_df

# 예시 실행
cleaned_age_exist = remove_outliers(age_exist, '현재 직업', '나이')
```

##### 목표

```python
# 목표 항목 추출
goal1 = das.query('월 == "11월" and `Q2.` != "응답없음"')['Q2.']
goal2 = das.query('월 == "12월" and `Q2.` != "응답없음"')['Q2.']

goal_combined = pd.concat([goal1, goal2], ignore_index=True)
```

```python
# 1. 불용어 로드
with open('stopword.txt', encoding='utf-8') as f:
    stopwords = {line.strip() for line in f if line.strip()}

# 2. 형태소 분석기
okt = Okt()

# 문장에서 명사만 추출하고 불용어/한글자 제거 후 공백으로 연결 함수
def extract_nouns(text: str) -> str:
    nouns = okt.nouns(text)
    filtered = [n for n in nouns if n not in stopwords and len(n) > 1]
    return ' '.join(filtered)

# 3. 대상 텍스트 준비 (문자열화 + 결측/공백 정리)
target_texts3 = [str(t).strip() for t in goal_combined if str(t).strip()]

# 4. 명사 추출 + 불용어 제거
processed_texts3 = [extract_nouns(text) for text in target_texts3]

# 5. 벡터화 (단어 빈도 행렬)
vectorizer = CountVectorizer()  # 필요시 min_df=2 등 하이퍼파라미터 조정 가능
X3 = vectorizer.fit_transform(processed_texts3)

# 6. 단어 빈도 합산
word_counts3 = X3.sum(axis=0)  # 1 x vocab_size
vocab = vectorizer.vocabulary_  # {단어: 인덱스}
word_freq3 = [(word, int(word_counts3[0, idx])) for word, idx in vocab.items()]


word_freq3.sort(key=lambda x: x[1], reverse=True)
df_word_freq3 = pd.DataFrame(word_freq3, columns=["단어", "등장횟수"])

df_word_freq3
```

```python
# 1. 상위 20개 단어 선택
top20 = df_word_freq3.head(20)

# 2. 가로 막대그래프 그리기
plt.figure(figsize=(10, 8))
plt.barh(top20['단어'], top20['등장횟수'], color='skyblue')
plt.xlabel('등장 횟수', fontsize=13)
plt.ylabel('단어', fontsize=13)
plt.title('이루고 싶은 목표 항목-상위 20개 단어', fontsize=16, fontweight='bold')

# y축 단어 텍스트 크기 키우기
plt.yticks(fontsize=15)


plt.gca().invert_yaxis()  # 가장 빈도 높은 단어가 위로 오도록
plt.tight_layout()
plt.show()
```
